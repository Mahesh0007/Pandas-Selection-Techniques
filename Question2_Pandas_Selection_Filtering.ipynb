## Create a pandas DataFrame using the employee data provided

```python
import pandas as pd

data = {
    'EmployeeID': [201, 202, 203, 204, 205, 206, 207, 208, 209, 210],
    'Name': ['Sarah', 'Michael', 'Jennifer', 'David', 'Lisa', 'James', 'Emma', 'Robert', 'Sophia', 'William'],
    'Department': ['Engineering', 'Sales', 'Marketing', 'Engineering', 'HR', 'Sales', 'Engineering', 'Marketing', 'Sales', 'HR'],
    'Salary': [85000, 62000, 71000, 92000, 58000, 68000, 78000, 75000, 65000, 60000],
    'Performance Score': [4, 3, 5, 5, 4, 4, 3, 4, 5, 3],
    'Years of Service': [5, 3, 7, 8, 4, 6, 2, 5, 4, 3]
}

df = pd.DataFrame(data)
df.set_index('EmployeeID', inplace=True)
print(df)
```

## Select and display only the 'Name', 'Department', and 'Salary' columns

```python
print(df[['Name', 'Department', 'Salary']])
```

## Use iloc to display the employee at row position 5

```python
print(df.iloc[5])
```

## Use loc to display the employee with EmployeeID 208

```python
print(df.loc[208])
```

## Filter and display all employees from the 'Engineering' department

```python
print(df[df['Department'] == 'Engineering'])
```

## Filter and display all employees with Salary greater than 70000

```python
print(df[df['Salary'] > 70000])
```

## Filter and display employees from 'Sales' department with Performance Score >= 4

```python
print(df[(df['Department'] == 'Sales') & (df['Performance Score'] >= 4)])
```

## Sort by Salary descending and display top 3 highest-paid employees

```python
print(df.sort_values('Salary', ascending=False).head(3))
```

## Sort by Department (ascending) and then by Salary (descending) within each department

```python
print(df.sort_values(by=['Department', 'Salary'], ascending=[True, False]))
```
